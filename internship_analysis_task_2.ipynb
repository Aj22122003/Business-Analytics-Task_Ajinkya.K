```python
import pandas as pd

# Load the dataset again (adjust the path if needed)
df = pd.read_csv(r'C:/Users/bhard/Desktop/DATA ANALYST OFFLINE/Python/internship/Dataset.csv')
```

```python
import matplotlib.pyplot as plt

# Gender Information
gender_counts = df['gender'].value_counts()

# gender counts
print("Gender Distribution:\n", gender_counts)

# Visualization: Bar Chart
plt.figure(figsize=(8, 6))
gender_counts.plot(kind='bar', color=['skyblue', 'orange', 'green'])
plt.title('Gender Distribution in Dataset', fontsize=16)
plt.xlabel('Gender', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(rotation=0, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Visualization: Pie Chart
plt.figure(figsize=(8, 6))
gender_counts.plot(kind='pie', autopct='%1.1f%%', colors=['skyblue', 'orange', 'green'], startangle=90)
plt.title('Gender Distribution in Dataset', fontsize=16)
plt.ylabel('')  
plt.tight_layout()
plt.show()
```
